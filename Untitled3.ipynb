```python
# Function to interpolate every 10 meters
def interpolate_coordinates(coords, distance=10):
  new_coords = [coords[0]]
  for i in range(1, len(coords)):
    start = coords[i-1]
    end = coords[i]
    segment_distance = geodesic(start, end).meters
    if segment_distance > distance:
      num_points = int(segment_distance // distance)
      for j in range(1, num_points + 1):
        lat = start[0] + j * (end[0] - start[0])/ num_points
        lng = start[1] + j * (end[1] - start[1])/ num_points
        new_coords.append((lat, lng))
  new_coords.append(coords[-1])
  return new_coords

new_coords = interpolate_coordinates(coords, distance=10)
new_coords, len(new_coords)
```

```python
def get_intrinsics(H, W, fov=60):
  f = 0.5 * W / np.tan(0.5 * fov * np.pi / 180.0)
  cx = 0.5 * W
  cy = 0.5 * H
  return np.array([[f, 0, cx],
                   [0, f, cy],
                   [0, 0, 1]])
```

```python
def pixel_to_point(depth_image, camera_intrinsics=None):
  height, width = depth_image.shape
  if camera_intrinsics is None:
    camera_intrinsics = get_intrinsics(height, width, fov=60)

  # Create u, v meshgrid and precompute projection triangle ratios
  fx, fy = camera_intrinsics[0, 0], camera_intrinsics[1, 1]
  cx, cy = camera_intrinsics[0, 2], camera_intrinsics[1, 2]

  x = np.linspace(0, width - 1, width)
  y = np.linspace(0, height - 1, height)
  u, v = np.meshgrid(x, y)

  x_over_z = (u - cx) / fx
  y_over_z = (v - cy) / fy

  # 3D Pythagoras re-arranged to solve for z
  z = depth_image / np.sqrt(x_over_z**2 + y_over_z**2 + 1)
  x = x_over_z * z
  y = y_over_z * z

  return x, y, z
```

```python
def create_point_cloud(depth_image, color_image, camera_intrinsics=None, scale_ratio=200.0):

  height, width, *_ = depth_image.shape  # Use *_ to ignore extra dimensions
  if camera_intrinsics is None:
    camera_intrinsics = get_intrinsics(height, width, fov=60)

  color_image = cv2.resize(color_image, (width, height))

  # make sure that depth_image does not contain zeros
  depth_image = np.maximum(depth_image, 1e-5)

  # Apply sky filtering
  #depth_image = filter_sky(color_image, depth_image)

  depth_image = scale_ratio * depth_image
  x, y, z = pixel_to_point(depth_image, camera_intrinsics)
  point_image = np.stack([x, y, z], axis=-1)

  cloud = o3d.geometry.PointCloud()
  cloud.points = o3d.utility.Vector3dVector(point_image.reshape(-1, 3))
  cloud.colors = o3d.utility.Vector3dVector(color_image.reshape(-1, 3) / 255.0)

  return cloud
```
